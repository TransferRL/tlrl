```python
for x in range(10000):
    state, reward, done, info = env.step(2) 
print(state)
plt.figure()
plt.imshow(env.render(mode='rgb_array'))
env.render(close=True)
```

```python
for x in range(10000):
    if np.isclose(state(2), 0) and state(0)<-5.2:
        state, reward, done, info = env.step(2)
    elif np.isclose(state(2), 0) and state(0)>-5.2:
        state, reward, done, info = env.step(1)
    elif state(2) > 0:
        state, reward, done, info = env.step(2)
    elif state(2) < 0:
        state, reward, done, info = env.step(1)
    else:
        state, reward, done, info = env.step(2)
        
        
print(state)
plt.figure()
plt.imshow(env.render(mode='rgb_array'))
env.render(close=True)
```
